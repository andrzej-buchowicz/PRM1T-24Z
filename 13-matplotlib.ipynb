### Wykres funkcji jednej zmienej

```python
import numpy as np

import matplotlib as mpl
import matplotlib.pylab as plt
```

```python
t = np.linspace(0., 1., num=100)
x = np.sin(2 * np.pi * t)
y = np.cos(2 * np.pi * t)
```

```python
_, ax = plt.subplots(figsize=(9, 6))
ax.plot(t, x, 'r-', label=r'$\sin(2 \pi t)$')
ax.plot(t, y, color='#000080', linestyle='dashed', label=r'$\cos(2 \pi t)$')
ax.grid()
ax.set_xlabel('t')
ax.set_ylabel('x(t)')
ax.set_title("Wykres funkcji")
ax.legend(loc='upper right')
#plt.show()

plt.savefig('wykres.png')
```

```python
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(9, 6))

fig.suptitle("Wykres funkcji")

ax1.plot(t, x, 'r-')
ax1.grid()
ax1.set_ylabel(r'$\sin(2 \pi t)$')

ax2.plot(t, y, color='#000080', linestyle='dashed')
ax2.grid()
ax2.set_xlabel('t')
ax2.set_ylabel(r'$\cos(2 \pi t)$');
```

```python
t = np.linspace(0, 0.5, 20)
x = np.sin(2 * np.pi * t)

_, ax = plt.subplots(figsize=(9, 6))
ax.stem(t, x);
```

### Wyniki pomiarw

```python
x_pom = np.linspace(0., 10, num=10)
y_pom = np.polyval([0.5, 1.0, 2.], x_pom) + 2.5 * np.random.randn(x_pom.shape[0])
```

```python
_, ax = plt.subplots(figsize=(9, 6))
ax.plot(x_pom, y_pom, marker='d')   

ax.grid()
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('Wyniki pomiarw');
```

```python
p = np.polyfit(x_pom, y_pom, 2)
p
```

```python
x = np.linspace(0, 10, num=100)
y = np.polyval(p, x)
```

```python
_, ax = plt.subplots(figsize=(9, 6))

ax.scatter(x_pom, y_pom, marker='d', color='#ff0000', label='pomiar')
ax.plot(x, y, color='#ff8000', linestyle='dashed', label='aprox')

ax.grid()
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.legend()
ax.set_title('Wyniki pomiarw');
```
